#### Without using estimator API:

```python
import numpy as np
import pandas as pd
import tensorflow as tf

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define data
x_data = np.linspace(0.0, 10.0, 1000000)
noise = np.random.randn(len(x_data)) # add noise

y_true = 0.5*x_data + 5 + noise

# Convert to pandas dataframes
X = pd.DataFrame(x_data, columns = ["X"])
y = pd.DataFrame(y_true, columns = ['y'])

my_data = pd.concat([X, y], axis=1)

# plot sample of the data
my_data.sample(n=250).plot(kind="scatter", x='X', y='y')
plt.show()
```

```python
# define batch size and 
batch_size = 8
# Initialize values for slope and intercept
m_ini, b_ini = np.random.randn(2)

# initialize the data (convert numpy float to python float)
m = tf.Variable(float(str(m_ini)))
b = tf.Variable(float(str(b_ini)))

# Create the placeholders
xph = tf.placeholder(tf.float32, [batch_size])
yph = tf.placeholder(tf.float32, [batch_size])

# define the model
y_model = m*xph + b
# define the loss function:
error = tf.reduce_sum(tf.square(yph-y_model))

# Use optimizer from TensorFlow and define the train object
# Setup optimizer and trainer
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.001)
train = optimizer.minimize(error)
```

```python
# Initialize the tensor flow session:
init = tf.global_variables_initializer()

# Run tf session with bath sizes
with tf.Session() as sess:
    sess.run(init)
    batches = 10000 # batches times batch_size = no. of operations
    for i in range(batches):
        rand_ind = np.random.randint(len(x_data), size=batch_size)
        feed = {xph:x_data[rand_ind], yph:y_true[rand_ind]}
        sess.run(train,  feed_dict=feed)
        
    model_m, model_b = sess.run([m, b])

# Predicted values
y_hat = x_data*model_m + model_b

# Check how well the predicted values fit the data
my_data.sample(250).plot(kind="scatter", x="X", y="y")
plt.plot(x_data, y_hat, "r")
plt.show()
```

#### By using estimator API (tf.estimator.<>):
- Linear: LinearClassifier, and LinearRegressor
- Densely connected: DNNClassifier, and DNNRegressor 
- Combined versions: DNNLinearCombinedClassifier, and DNNLinearCombinedRegressor

#### List of task
1. Define a list of feature columns
2. Create the estimator model
3. Create the data input function
4. Call train, evaluate, and predict methods on the estimator object

```python
from sklearn.model_selection import train_test_split
```

```python
# setup a feature column (step 1)
fea_cols = [tf.feature_column.numeric_column('x', shape=[1])]
```

```python
# setup an estimator (step 2) - similar to sklearn estimator*
estimator = tf.estimator.LinearRegressor(feature_columns=fea_cols)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_true, 
                                                    test_size = 0.3,
                                                    random_state=101)
```

```python
# setup the input function (step 3)

input_fun = tf.estimator.inputs.numpy_input_fn({'x': X_train},
                                               y_train, batch_size=8,
                                               num_epochs=None,
                                               shuffle=True)

train_input_fun = tf.estimator.inputs.numpy_input_fn({'x': X_train},
                                                     y_train, batch_size=8,
                                                     num_epochs=1000,
                                                     shuffle=False)

test_input_fun = tf.estimator.inputs.numpy_input_fn({'x': X_test},
                                                     y_test, batch_size=8,
                                                     num_epochs=1000,
                                                     shuffle=False)
```

```python
estimator.train(input_fn=input_fun, steps=1000)
```

```python
train_metrics = estimator.evaluate(input_fn=train_input_fun, steps=1000)
```

```python
test_metrics = estimator.evaluate(input_fn=test_input_fun, steps=1000)
```

```python
# Pay attention to 'loss', should be comparable and low
# in both train and test metric

print("Training metric:")
print(train_metrics)
print("Evaluation metric:")
print(test_metrics)
```

```python
new_data = np.linspace(0, 10, 10)
pred_input_fun = tf.estimator.inputs.numpy_input_fn({'x': new_data},
                                                    shuffle=False)

list(estimator.predict(input_fn=pred_input_fun))
```

```python
predictions = []

for pred in estimator.predict(input_fn=pred_input_fun):
    predictions.append(pred['predictions'])
```

```python
my_data.sample(n=250).plot(kind="scatter", x="X", y='y')
plt.plot(new_data, predictions, 'r*')
```
